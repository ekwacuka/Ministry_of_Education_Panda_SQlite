```python
import sqlite3

## Creates a database connection SQLite and creates a database.

conn = sqlite3.connect('MINISTRYOFEDUCATION.db')

## Create a Cursor Object 

cursor = conn.cursor()

print("\t MINISTRYOFEDUCATION Database created \n")

## Creates Teachers table

cursor.execute(''' CREATE TABLE IF NOT EXISTS teacher (
        COUNTY VARCHAR(100),
        SCHOOL_TYPE VARCHAR(100),
        EMPLOYMENT_BODY VARCHAR(100),
        NO_OF_TEACHERS INT(15),
        YEAR DATETIME);''')

print("\t TEACHER TABLE STRUCTURE \n")

conn.commit()
```

```python
## writes records stored in a DataFrame to a SQLite database (defined in conn)

df.to_sql('teacher', conn, if_exists='replace', index=False)
```
